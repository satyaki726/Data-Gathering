# WEB SCRAPING

```python
final=pd.DataFrame()
for j in range(1,333):
    headers = {'User-Agent': 'Mozilla/5.0 (Wnidows NT 6.3; Win64; x64) AppleWebkit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage=requests.get('https://www.ambitionbox.com/list-of-companies?page={}'.format(j), headers=headers).text
    soup=BeautifulSoup(webpage,'lxml')
    company=soup.find_all('div',class_='company-content-wrapper')
    name=[]
    rating=[]
    reviews=[]
    tags=[]
    ctype=[]
    hq=[]
    old=[]
    employees=[]
    for i in company:
        name.append(i.find('h2').text.strip())
        rating.append(i.find('p',class_='rating').text.strip())
        reviews.append(i.find('a',class_='review-count').text.strip())
        tags.append(i.find('div',class_='chips-block').text.strip())
        ctype.append(i.find_all('p',class_='infoEntity')[0].text.strip())
        hq.append(i.find_all('p',class_='infoEntity')[1].text.strip())
        old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
        try:
            employees.append(i.find_all('p',class_='infoEntity')[3].text.strip())
        except:
            employees.append(np.nan)
    d={'Name':name,'Rating':rating,'Review':reviews,'Type':ctype,'Head_Quarter':hq,'Old':old,'Employees':employees,'Tags':tags}
    df=pd.DataFrame(d)
    final=final.append(df,ignore_index=True)
final    
```

```python
final.to_csv('Company_details.csv')
```
